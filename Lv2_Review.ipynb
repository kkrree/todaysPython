#### Data preprocessing
결측치 처리

```python
#결측치 제거
train.dropna(inplace=True)
#결측치 상수값으로 대체
test = test.fillna(0)
#결측치 평균값으로 대체
train.fillna(train.mean(),inplace=True)
#결측치 보간법으로 대체
train.interpolate(inplace=True)
```

```python
######inplace option#####
train.interpolate(inplace=True)
# train = train.interpolate() 
# 
# inplace option을 true로 주면 해당 메모리 주소를 그대로 이용하여, 원본을 대체해버린다. 
# 메모리를 재활용하는 방식으로 변수를 선언하여 해당 값에 inplace=true 옵션을 주면 메모리 주소상에 충돌이 발생하여 에러가 날 수 있다.
# 때문에 inplace = False일 경우 객체에 직접 선언하여 담아줄 수 있고, (원본의 값이 변경될 경우 수정된 객체는 원본과 다른 주소값을 가짐, 서로 다른 주소값을 갖는 객체의 == 화 가능)
# inplace = True 일 경우, 추가적으로 객체에 담아줄 필요 없이 해당 데이터를 그대로 사용하면 된다.

#test = test.fillna(0, inplace=True)
# 위와 같은 경우 
# AttributeError: 'NoneType' object has no attribute 'info' 
# 해당 에러가 날 수 있다.

train = train
test = test
#기본 df 객체끼리는 주소값 충돌이 나지 않는 것 같다..?
```

```python
print('tarin_isnull?\n',train.isnull().sum())
print('-----------------------------')
print('test_isnull?\n',test.isnull().sum())
```

#### Modeling

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(criterion='mse')

X_train = train.drop(['count'], axis=1)
Y_train = train['count']

model.fit(X_train, Y_train)
```

**| drop vs. dropna**   
When using drop only, it drops the columns/rows you define   
When using dropna, it removes all entries with NaN values (or null in general)

변수 중요도 확인

```python
model.feature_importances_
```

중요도 낮은 변수 제거

```python
train.info()
```

```python
X_train_v1 = X_train.drop(['id','hour_bef_precipitation','hour_bef_pm2.5'],axis=1)
test_v1 = test.drop(['id','hour_bef_precipitation','hour_bef_pm2.5'],axis=1)
```

```python
model_v1 = RandomForestRegressor(criterion='mse',random_state=2020)
```

#### tunning
**gridSearch** 하이퍼파라미터 튜닝(과적합 방지를 위한 정지규칙 활용) 중 완전 탐색방법


```python
from sklearn.model_selection import GridSearchCV

param = {'n_estimators' : [50,100,250],
         'max_depth' : [5,6,7],
         'min_samples_split' : [30,40,50]}

gs = GridSearchCV(estimator=model_v1, param_grid=param, scoring='neg_mean_squared_error',cv=3)

gs.fit(X_train_v1,Y_train)
```

#### 제출 파일 생성

```python
sub = pd.read_csv('data/bike/submission.csv')

pred = gs.predict(test_v1)

sub['count'] = pred

sub.to_csv('gridSearch.csv',index=False)
```
